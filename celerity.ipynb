# lowercasing all values in dataframe

```python
for i in dfexp1.columns:
    dfexp1[i]=dfexp1[i].str.lower()
```

# Transforming categorical columns to numerical 

```python
transformer=ColumnTransformer(transformers=[('tnf1',OneHotEncoder(sparse=False,drop='first'),['Origin Country','Destination Country'])],remainder='passthrough')
transformer.set_output(transform='pandas')
```

```python
ndf=dfexp1.iloc[:,0:6]
```

```python
ndf.head()
```

```python
newdata=transformer.fit_transform(ndf)
```

# Applying Decision Tree as model

```python
clf=DecisionTreeClassifier()
clf.fit(newdata,dfexp1.iloc[:,6])
```

# prediction for a test input

```python

test_input = pd.DataFrame({'Origin Country':['botswana'], 'Destination Country':['dr congo'], 'Quantity':[0], 'No Of Packages':[36],
       'Gross Weight':[35], 'Freight BWP':[5352]})
transinput=transformer.transform(test_input)
clf.predict(transinput)[0]
```

# Trade service routes of different freight forwardders

```python
routelist={}
for i in dfexp1['Declarant'].unique():
    ori=list(dfexp1['Origin Country'][dfexp1['Declarant']==i])
    des=list(dfexp1['Destination Country'][dfexp1['Declarant']==i])
    if i not in routelist:
        routelist[i]=[]
    for j in range(len(ori)):
            routelist[i].append((ori[j],des[j]))
            routelist[i]=list(set(routelist[i]))
```
